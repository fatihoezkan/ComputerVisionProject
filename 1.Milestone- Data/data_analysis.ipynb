```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Function to analyze an image
def analyze_image(image_path):
    # Load the image
    img = Image.open(image_path)

    # Convert the image to grayscale
    img_gray = img.convert('L')

    # Convert the grayscale image to a numpy array
    img_array = np.array(img_gray)

    # Calculate statistics
    min_intensity = np.min(img_array)
    max_intensity = np.max(img_array)
    mean_intensity = np.mean(img_array)
    std_intensity = np.std(img_array)

    # Return the calculated statistics
    return min_intensity, max_intensity, mean_intensity, std_intensity

# Define the directory where your .tiff images are stored
directory_path = '/Users/fatihozkan/Desktop/Semester_4/ComptVision/data/AgNors'

# Loop through all .tiff files in the directory and analyze them
for filename in os.listdir(directory_path):
    if filename.endswith('.tiff'):
        file_path = os.path.join(directory_path, filename)
        
        # Analyze the image
        stats = analyze_image(file_path)
        print(f"Image: {filename}")
        print(f"  Min intensity: {stats[0]}")
        print(f"  Max intensity: {stats[1]}")
        print(f"  Mean intensity: {stats[2]}")
        print(f"  Std intensity: {stats[3]}\n")


# Lists to store the statistics for all images
min_intensities = []
max_intensities = []
mean_intensities = []
std_intensities = []
image_names = []

# Loop through all .tiff files in the directory and analyze them
for filename in os.listdir(directory_path):
    if filename.lower().endswith('.tiff'):
        file_path = os.path.join(directory_path, filename)
        
        # Analyze the image
        stats = analyze_image(file_path)
        
        # Append the statistics to the lists
        min_intensities.append(stats[0])
        max_intensities.append(stats[1])
        mean_intensities.append(stats[2])
        std_intensities.append(stats[3])
        image_names.append(filename)  # Append the filename or a custom identifier

# # Now, plot the statistics
# x = range(len(image_names))  # Create a list of numbers for x-axis

# # Plotting all statistics on separate subplots
# fig, axs = plt.subplots(4, 1, figsize=(10, 20))

# # Min intensity plot
# axs[0].bar(x, min_intensities, color='gray')
# axs[0].set_title('Minimum Intensities')
# axs[0].set_xticks(x)
# axs[0].set_xticklabels(image_names, rotation='vertical')

# # Max intensity plot
# axs[1].bar(x, max_intensities, color='red')
# axs[1].set_title('Maximum Intensities')
# axs[1].set_xticks(x)
# axs[1].set_xticklabels(image_names, rotation='vertical')

# # Mean intensity plot
# axs[2].bar(x, mean_intensities, color='green')
# axs[2].set_title('Mean Intensities')
# axs[2].set_xticks(x)
# axs[2].set_xticklabels(image_names, rotation='vertical')

# # Std intensity plot
# axs[3].bar(x, std_intensities, color='blue')
# axs[3].set_title('Standard Deviation of Intensities')
# axs[3].set_xticks(x)
# axs[3].set_xticklabels(image_names, rotation='vertical')

# # Set up the layout and display the plot
# plt.tight_layout()
# plt.show()
```

```python
import matplotlib.pyplot as plt

# Data: Labels and their respective counts
labels = range(0, 12)  # Labels from 0 to 11
counts = [4461, 8152, 4003, 1628, 734, 356, 193, 78, 47, 28, 26, 28]  # The count for each label

# Creating the histogram
plt.figure(figsize=(10, 6))  # Set the figure size for better readability
plt.bar(labels, counts, color='skyblue')  # Create a bar chart with a custom color

# Adding title and labels
plt.title('Distribution of Annotations per Label', fontsize=15)
plt.xlabel('Label', fontsize=12)
plt.ylabel('Number of Annotations', fontsize=12)

# Adding the count above each bar for better clarity
for i in range(len(labels)):
    plt.text(x=labels[i]-0.25, y=counts[i]+100, s=str(counts[i]), size=10)

# Adjusting the x-axis to show each label
plt.xticks(labels)

# Show the plot
plt.tight_layout()
plt.show()
```
